Source: https://data.wprdc.org/dataset/pbp-fire-arm-seizures

## Notebook Setup
As always, start out with some imports!

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Preparing the DataFrame

To start out, we are going to quickly load in this data set.
If necessary, some cleaning operat

```python
firearm_data = pd.read_csv("firearm_seizures.csv")

# some entries don't have data on the neighborhood!
firearm_data = firearm_data.dropna()
```

## Analysis

This data set is pretty nice, which means we are ready to plot our values.

In our case, we want these values to be ascending, as the neighborhood **most** friendly to a mob boss would have the **least** number of firearm seiures.

```python
firearm_data.groupby("neighborhood").sum()["total_count"].sort_values(ascending=True).plot.bar(colormap='Accent', figsize=(28, 12)).set_ylabel("total firearm seizures")
plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='medium',
    fontsize='medium',
)
```

## Conclusion

Based on firearm seizures alone, we can conclude that there are several top neighborhoods for a mob boss.

This is exactly why we will take a look at some other statistics to determine the best one!
