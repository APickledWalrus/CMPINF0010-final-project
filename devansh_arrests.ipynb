# loading pandas module and importing data

```python
# load pandas
import pandas as pd
import numpy as nm
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# load data

police_arrests_pandas = pd.read_csv("arrests.csv")
police_arrests_pandas
```

```python
# How many rows and columns
police_arrests_pandas.shape
```

```python
#counting number of arrests for each area
police_arrests_pandas['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)
```

# Mt. Oliver neighborhood has the least number of arrests while Central business district has the highest number of arrests


```python
# HOW MANY NEIGHBORHOODS ARE THERE IN PITTSBURG
data= police_arrests_pandas['INCIDENTNEIGHBORHOOD'].unique()
data

#len(police_arrests_pandas['INCIDENTNEIGHBORHOOD'].unique())
```

```python
police_arrests_pandas.groupby("INCIDENTNEIGHBORHOOD").count()
police_arrests_pandas["INCIDENTNEIGHBORHOOD"].value_counts().plot(kind="bar", figsize=(28,12))
plt.xticks(
    rotation = 45,
    horizontalalignment="right",
    fontweight="medium",
    fontsize="large"
)
```
